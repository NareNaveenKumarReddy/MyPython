### Logical Operators

```python
print(10 > 5 and 10 < 20) # and means both the conditions must be true
print(10 < 5 and 10 < 20) # true and false
```

```python
print(10 > 5 or 10 < 20) # if both the conditions are false it would give me false
print(10 > 5 or 10 > 20) # else in all other cases it would give me true
```

```python
print(not(10 > 5 and 10 < 20))
print(not(10 < 5 and 10 < 20))
print(not(True)) # not convert true to false and false to true
```

## Conditionals and Indentations

### if

```python
a = 10
if a == 10:
    print(a)
```

### if/else

```python
a = 100
if a == 10:
    print("Equal")
else:
    print("not equal")
```

```python
a = 10
if a == 10:
    print("Equal")
    a = 100
    print(a)
else:
    print("not equal")
```

### elif

```python
a = 100
if a == 10:
    print("Equal")
elif a > 10:
    print("Greater")
else:
    print("Smaller")
```

```python
# if statement in one line
a = 10
if a == 10: print("equal")
```

```python
# if/else statement in one line
a = 5
print("equal") if a == 10 else print("Not equal")
```

```python
a = 10
print("less") if a < 10 else print("greater") if a > 10 else print("Equal")
```

```python
a = 10
b = 15
if a == 10 and b < 12:
    print("hello")
```

```python
a = 10
b = 2
if a != 10 or b > 12:
    print("hello")
```

### Nested if

```python
a = 2
if a > 10:
    print("greater than 10")
    if a > 50:
        print("greater than 50")
        if a > 60:
            print("greater than 60")
else:
    print("less")
```

### pass

```python
a = 3
if a < 10:
    print("less than 10")
else:
    pass
```

## Loops

### For loops

```python
list_of_numbers = [10, 20 ,30 ,89]
```

```python
for i in list_of_numbers:
    print(i)
print("loop completed")
```

```python
for i in list_of_numbers:
    print(i)
    i += 10
    print(i)
    print("***********")
```

```python
tuple_of_numbers = (10, 20 ,30 ,89)
```

```python
for i in tuple_of_numbers:
    print(i)
```

```python
for i in tuple_of_numbers:
    print(i)
    i += 10
    print(i)
    print("***********")
```

```python
set_of_numbers = {10, 15, 19}
for i in set_of_numbers:
    print(i)
```

```python
new_dict = {
    "name": "Shreyas",
    "Gender": "Male",
    "Occupation": "AI Engineer",
    "Years of exp": 5
}

for i in new_dict:
    print(i)
```

```python
for i in new_dict.keys():
    print(i)
```

```python
for i in new_dict.values():
    print(i)
```

```python
for k,v in new_dict.items():
    print("{} : {}".format(k,v))
```

```python
for i in range(5,10):
    print(i)
```

## break, continue and else

```python
x = 1
while x < 10:
    x += 1
    if x == 5:
        break
    print(x)
```

```python
a = [10, 20, 30]
for i in a:
    if i == 20:
        break
    print(i)
```

```python
x = 1
while x < 10:
    x += 1
    if x == 5:
        continue
    print(x)
```

```python
a = [10, 20, 30]
for i in a:
    if i == 20:
        continue
    print(i)
```

```python
x = 1
while x < 10:
    x += 1
    print(x)
else:
    print("Not less than 10")
```

```python
a = [10, 20, 30]
for i in a:
    print(i)
else:
    print("Completed the loop")
```
