```python
lang = ['Java','C','Python','R']
trend17 = [10,7,21,17]
trend18 = [6,4,15,11]
plt.bar(lang,trend17,label='2017')
plt.bar(lang,trend18,label='2018')
plt.xlabel('Programming Languages')
plt.ylabel('Trending rate')
plt.title("Trend of programming languages in DS Field")
plt.legend()
plt.grid()
```

```python
lang = ['Java','C','Python','R']
trend17 = [10,7,21,17]
trend18 = [6,4,15,11]
plt.barh(lang,trend17,label='2017')
plt.barh(lang,trend18,label='2018')
plt.xlabel('Programming Languages')
plt.ylabel('Trending rate')
plt.title("Trend of programming languages in DS Field")
plt.legend()
plt.grid()
```

```python
lang = ['Java','C','Python','R']
xpos = np.arange(len(lang))
print(xpos)
trend17 = [6,4,15,11]
trend18 = [10,7,21,17]
plt.xticks(xpos, lang)
plt.bar(xpos-0.2,trend17, width = 0.4,label='2017')
plt.bar(xpos+0.2,trend18,width = 0.4,label='2018')
plt.xlabel('Programming Languages')
plt.ylabel('Trending rate')
plt.title("Trend of programming languages in DS Field")
plt.legend()
```

```python
categories = ['Study','Social media','Work','Exercise','Sleep']
time_spent = [2,4,8,1,9]
```

```python
plt.pie(time_spent,labels = categories,radius=2)
plt.show()
```

```python
plt.pie(time_spent,labels = categories,radius=2, autopct = '%0.2f%%', shadow=True)
plt.show()
```

```python
plt.pie(time_spent,labels = categories,radius=2, autopct = '%0.2f%%', shadow=True, explode=[0,0,0.8,0,0.3])
plt.show()
```

```python
plt.pie(time_spent,labels = categories,radius=2, autopct = '%0.2f%%', shadow=True, explode=[0,0,0.3,0,0])
plt.savefig('my_pie.png')
```

```python
plt.pie(time_spent,labels = categories,radius=2, autopct = '%0.2f%%', pctdistance=1.1, labeldistance=0.6,shadow=True, explode=[0,0,0.3,0,0])
plt.savefig('my_pie.png', bbox_inches = 'tight', facecolor = 'white')
```

```python
# multiplots

x = np.arange(0, 10, 0.01)
  
plt.subplot(3, 1, 1) 
plt.plot(x, np.sin(x)) 
  
plt.subplot(3, 1, 2) 
plt.plot(x, np.cos(x)) 
  

plt.subplot(3, 1, 3) 
plt.plot(x, x) 
  
plt.show() 
```

```python
x = np.arange(0, 10, 0.01)
  
plt.subplot(2, 2, 1) 
plt.plot(x, np.sin(x)) 
  
plt.subplot(2, 2, 2) 
plt.plot(x, np.cos(x)) 
  
plt.subplot(2, 2, 3) 
plt.plot(x, x)

plt.subplot(2, 2, 4)
plt.hist(x, bins= [0,0.2,0.4,0.6,0.8,1])
  
plt.show() 
```
